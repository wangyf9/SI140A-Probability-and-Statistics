Question1（1）  
    In this question, $F(X)=e^x/(1+e^x)$, then we can transform it and get $F^{-1}(X)=log(x/(1-x))$.  
    Because F(X) is continuous and increasing, then in $F^{-1}(X)$, X is $\in$ (0,1).  
    Then we can replace X with $U\sim Unif(0,1)$, and get $F^{-1}(U)=log(u/(1-u))$.

```python
u=np.random.random(100000)
plt.figure(1)
hist_1_1=plt.hist(np.log(u/(1-u)),bins=100,edgecolor="black",density=True)
x=hist_1_1[1][0:len(hist_1_1[1])-1]
y=hist_1_1[0]
plt.plot(x,y,color='r')
plt.xlabel("X")
plt.ylabel("frequency")
plt.title("Logistic distribution")
plt.show()
```

Question 2(2)  
Binomial distribution, and n=20, p=0.5, $p(X=k)=\binom {20} {k}p^k(1-p)^{20-k}$.

```python
x=np.zeros(100000)
x1=np.zeros(21)
x_axis=np.arange(21)
def Binomial():
    sum=0
    gen=np.random.randint(low=0,high=2,size=20)
    for i in range(20):
        if gen[i]==1:
            sum+=1
    return sum
for i in range(100000):
    x1[Binomial()]+=1
x1=x1/100000  
plt.bar(x_axis,x1)
plt.plot(x_axis,x1,color='r')
plt.xlabel("X")
plt.ylabel("frequency")
plt.title("Binomial distribution Bin(20,0.5)")
plt.show()
```

Question 2(3)  
Geometric distribution, p=0.5, $p(X=k)=p(1-p)^{k}$.

```python
col=np.zeros(51)
x_axis=np.arange(51)
def Geom():
    index=0
    gen=np.random.randint(low=0,high=2,size=50)
    for i in range(50):
        if gen[i]==0:
            index+=1
        else:
            break
    return index
for i in range(100000):
    col[Geom()]+=1
col=col/100000  
plt.bar(x_axis,col)
plt.plot(x_axis,col,color='r')
plt.xlabel("X")
plt.ylabel("frequency")
plt.title("Geometric distribution Geom(0.5)")
plt.show()
```

Question 2(4)  
Negative Binomial distribution, r=10, p=0.5, $p(X=k)=\binom {k+r-1}{k}p^{r}(1-p)^{k}$.

```python
col=np.zeros(51)
x_axis=np.arange(51)
def NBin():
    num_fai=0
    num_suc=0
    gen=np.random.randint(low=0,high=2,size=50)
    for i in range(50):
        if gen[i]==0:
            num_fai+=1
        elif gen[i]==1:
            num_suc+=1
        if num_suc==10:
            break
    return num_fai
for i in range(100000):
    col[NBin()]+=1
col=col/100000  
plt.bar(x_axis,col)
plt.plot(x_axis,col,color='r')
plt.xlabel("X")
plt.ylabel("frequency")
plt.title("Negative binomial distribution Geom(0.5)")
plt.show()
```

Question3(4)  
From $\prod_{i=1}^nU_i>=e^{-1}$, we can transform it as $\sum_{i=1}^nlogU_i>=-1$  
Because $U_i\sim Unif(0,1)$, then $logU_i$ has the CDF $e^{-x}$ where $x>=0$  
Let $X_i$ be $-logU_i$, then $X_i$ has the CDF $1-e^{-x}$ $x>=0$  
Then we can get the PDF $e^{-x}$, then $X_i\sim Expo(1)$.  
Then let $X=X_1+X_2+\cdots+X_n$, then the PDF of X is $e^{-x}/n!$.  
$P(N=k)=P(N>k)-P(N>k-1)$  
$=P(X_1+X_2+\cdots+X_k<1)-P(X_1+X_2+\cdots+X_{k-1}<1)$  
$=\int_{-\infty}^1 e^{-x}/k!dx-\int_{-\infty}^1 e^{-x}/(k-1)!dx$  
$=e^{-1}/k!$  
Also when n=0, it satisfies too.  
Therefore, $N\sim Pois(1)$

```python
car=np.random.randint(low=1,high=4,size=100000)
choose=np.random.randint(low=1,high=4,size=100000)
win_stay=0
win_switch=0
def Monty(i):
    for j in range(1,4):
        if j!=car[i] and j!=choose[i]:
            return j
def switch(i,Mon):
    for j in range(1,4):
        if j!=choose[i] and j!=Mon:
            return j
for i in range (100000):
    Mon=Monty(i)
    if car[i]==choose[i]:
        win_stay+=1
    if switch(i,Mon)==car[i]:
        win_switch+=1
win_stay=win_stay/100000
win_switch=win_switch/100000
print("P(stay and win)=",win_stay)
print("P(switch and win)=",win_switch)
```

$P(stay and min)\approx 1/3$  
$P(switch and min)\approx 2/3$

```python
n1=4
n2=100
#(1)
def stay_without_switch(num):
    car=np.random.randint(low=0,high=num,size=100000)
    choose=np.random.randint(low=0,high=num,size=100000)
    sum=0
    for i in range(100000):
        if car[i]==choose[i]:
            sum+=1
    return sum/100000
print('P(stay when n=4)=',stay_without_switch(n1))
print('P(stay when n=100)=',stay_without_switch(n2))
```

$P(stay\ when\ n=4)\approx 0.25$  
$P(stay\ when\ n=100)\approx 0.01$

```python
#(2)
def always_switch(num):
    sum=0
    for i in range(5000):
        car=np.random.randint(1,num+1)
        rest=np.arange(1,num+1)
        choose=np.random.randint(1,num+1)
        Monty_rest_choose=np.arange(1,num+1)
        Monty_rest_choose=np.delete(Monty_rest_choose,np.where(Monty_rest_choose==car))
        n=num
        while n>2:
            n-=1
            inter_rest_Monty=np.delete(Monty_rest_choose,np.where(Monty_rest_choose==choose))
            Mon_really_choose=np.random.choice(inter_rest_Monty)
            Monty_rest_choose=np.delete(Monty_rest_choose,np.where(Monty_rest_choose==Mon_really_choose))
            rest=np.delete(rest,np.where(rest==Mon_really_choose))
            rest_inter=np.delete(rest,np.where(rest==choose))
            choose=np.random.choice(rest_inter)
        if car==choose:
            sum+=1
    return sum/5000
print('P(switch when n=4)=',always_switch(n1))
print('P(switch when n=100)=',always_switch(n2))
```

take one of answers for example  
P(switch when n=4)=0.6266  
P(switch when n=100)=0.64

```python
#(2)
def stay_last_switch(num):
    sum=0
    for i in range(5000):
        car=np.random.randint(1,num+1)
        rest=np.arange(1,num+1)
        choose=np.random.randint(1,num+1)
        Monty_rest_choose=np.arange(1,num+1)
        Monty_rest_choose=np.delete(Monty_rest_choose,np.where(Monty_rest_choose==car))
        n=num
        while n>2:
            n-=1
            inter_rest_Monty=np.delete(Monty_rest_choose,np.where(Monty_rest_choose==choose))
            Mon_really_choose=np.random.choice(inter_rest_Monty)
            Monty_rest_choose=np.delete(Monty_rest_choose,np.where(Monty_rest_choose==Mon_really_choose))
            rest=np.delete(rest,np.where(rest==Mon_really_choose))
        rest_final=np.delete(rest,np.where(rest==choose))
        choose=np.random.choice(rest_final)
        if car==choose:
            sum+=1
    return sum/5000
print('P(switch when n=4)=',stay_last_switch(n1))
print('P(switch when n=100)=',stay_last_switch(n2))
```

```python
def init(size):##initialize
    global father
    father=np.arange(size*size+2)
def find(x):##find father node
    if(father[x]==x):
        return x
    else:
        return find(father[x])
def merge(x,y):##connect
    father[find(x)]=find(y)
def percolate(n):#work
    grid =np.zeros([n,n])##initialize grid
    sum=0
    init(n)
    for i in range(n):
        merge(i,n*n)
        merge(n*n+1,i+n*(n-1))
    while(find(n*n+1)!=find(n*n)):
        sum+=1
        x=np.random.randint(0,n)
        y=np.random.randint(0,n)
        while(grid[x,y]!=0):#alright full, run towards another grid point
            x=np.random.randint(0,n)
            y=np.random.randint(0,n)
        grid[x,y]=1#full
        z=x*n+y#real postion
        ##check around and connect
        if 0<=x<n and 0<=y-1<n and grid[x,y-1]==1:
            merge(z,z-1)
        if 0<=x<n and 0<=y+1<n and grid[x,y+1]==1:
            merge(z,z+1)
        if 0<=x-1<n and 0<=y<n and grid[x-1,y]==1:
            merge(z,z-n)
        if 0<=x+1<n and 0<=y<n and grid[x+1,y]==1:
            merge(z,z+n)
    return sum
sum1=[]
sum2=[]
sum3=[]
for i in range(2000):
    sum1.append(percolate(20))
    sum2.append(percolate(50))
    sum3.append(percolate(100))
sum1=np.array(sum1)/400
sum2=np.array(sum2)/2500
sum3=np.array(sum3)/10000
print(f"When n = 20,mean = {sum1.mean()},variance = {sum1.var()}")
print(f"When n = 50,mean = {sum2.mean()},variance = {sum2.var()}")
print(f"When n = 100,mean = {sum3.mean()},variance = {sum3.var()}")
```

When n = 50, mean = 0.59293, variance = 0.00070  
When n = 100, mean = 0.59271, variance = 0.00026
